# Visualize Dataset

Set dataset path.

```python
DATASET_PATH = "dataset/"
```

```python
from PIL import Image, ImageDraw
import os
import json

def convertToAbsPoints(pos, size, totalSize):
    return [(pos["x"] * totalSize[0],
             pos["y"] * totalSize[1]),
            
            ((pos["x"] + size["x"]) * totalSize[0],
             (pos["y"] + size["y"]) * totalSize[1])]

file_index = 0
while os.path.isfile(DATASET_PATH + str(file_index) + ".png") and os.path.isfile(DATASET_PATH + str(file_index) + ".json"):
    print(f"Image {file_index}:")
    with Image.open(DATASET_PATH + str(file_index) + ".png") as im:
        with open(DATASET_PATH + str(file_index) + ".json") as json_file:
            annotations = json.load(json_file)
            for data in annotations.values():
                draw = ImageDraw.Draw(im)
                draw.rectangle(convertToAbsPoints(data["pos"], data["size"], im.size), outline="red")
                
        display(im)
    
    file_index = file_index + 1
            
```
