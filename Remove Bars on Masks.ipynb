```python
# Path of directory at which the dataset is saved.
DATASET_PATH = "dataset/"

status_bar_height = 98
nav_bar_height = 44
```

```python
from PIL import Image, ImageDraw
import os

file_index = 0
while os.path.isfile(DATASET_PATH + str(file_index) + ".json"):
    mask_index = 0
    
    while os.path.isfile(DATASET_PATH + str(file_index) + "_" + str(mask_index) + ".png"):
        with Image.open(DATASET_PATH + str(file_index) + "_" + str(mask_index) + ".png") as im:
            width, height = im.size
            draw = ImageDraw.Draw(im)
            draw.rectangle(((0, 0), (width, status_bar_height)), fill="white")
            draw.rectangle(((0, height), (width, height - nav_bar_height)), fill="white")

            display(im)
            im.save(DATASET_PATH + str(file_index) + "_" + str(mask_index) + ".png")

        mask_index = mask_index + 1
    
    file_index = file_index + 1
```
